```python
import requests
import pandas as pd
from bs4 import BeautifulSoup


data = {}

for page_no in range(0, 11):

    URL = f"https://quotes.toscrape.com/page/{page_no + 1}/"
    res = requests.get(URL)

    if res.status_code == 200:
        file_name = f"scarped_data/page{page_no + 1}.html"

        # save html
        with open(file_name, "w", encoding="utf-8") as f:
            f.write(res.text)

        # read html
        with open(file_name, "r", encoding="utf-8") as f:
            soup = BeautifulSoup(f.read(), "lxml")

        quotes_divs = soup.find_all("div", class_="quote")

        for i in quotes_divs:
            all_tags = i.find("div", class_="tags").get_text().lower()

            if "life" in all_tags:
                text = i.find("span", class_="text").get_text()
                author = i.find("small", class_="author").get_text()

                if author not in data:
                    data[author] = []
                data[author].append(text)

    else:
        print(f"Page {page_no + 1}: Data not found!")

# final output
for author, quotes in data.items():
    for quote in quotes:
        print(f"Author: {author}\nQuote: {quote}\n")
```

```python
rows = []

for author, quotes in data.items():
    for quote in quotes:
        rows.append({
            "author": author,
            "quote": quote
        })
```

```python
import pandas as pd
df = pd.DataFrame(rows)
print(df.head())
```

```python
df.to_csv("life_quotes.csv", index=False, encoding="utf-8")
```

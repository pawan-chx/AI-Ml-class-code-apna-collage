```python
df = pd.read_csv("Titanic-Dataset.csv")

# df.describe()
df["Survived"].count

```

```python
q2_a = df[["Name", "Sex", "Age", "Fare", "Survived"]]
q2_a
```

```python
q2_b = df[ (df["Sex"] == "female") & (df["Fare"] > 30) ]

q2_b
```

```python
sur = df["Survived"]

def cal_survival_rate(sur):
    live_count = 0 
    for ans in sur:
        if ans == 1:
           live_count += 1
        else:
            continue
    
    tot_per = len(df["Survived"])
    return (f"{(live_count/tot_per) * 100}%")        

print(cal_survival_rate(sur))
```

```python
q2_c = df.groupby("Pclass").agg(
    Survival_Rate_Percent=("Survived", lambda x: x.mean() * 100),
    Average_Fare=("Fare", "mean"),
    Average_Age=("Age", "mean")
)

q2_c

```
